# Matplotlib Work

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
sem = [1,2,3,4,5,6,7,8]
per = [80,82,52,68,77,79,86,45]
```

```python
plt.plot(sem, per, color='red', linewidth=3, linestyle='dashed')
plt.xlabel('Semester')
plt.ylabel('Percentage')
plt.title('Acad Record')
```

```python
plt.plot(sem, per, 'bo--')
```

```python
plt.plot(sem, per, marker = 'o', markersize = '10', linestyle = '')
```
